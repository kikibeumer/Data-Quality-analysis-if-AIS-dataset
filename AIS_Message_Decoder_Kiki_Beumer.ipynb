Binary to Float

```python
from ast import literal_eval

def binary_to_float(float_str):

    if (float_str)[0] == '-':
        float_str = f"-0b{float_str[1:]}"
    else :
        float_str = f"0b{float_str[1:]}"
    
    result = float(literal_eval(float_str))

    return result
    
```

Get destination in UN/LOCODE and ERI terminal codes

```python
def get_destination(binary_str):
    # AIS 6-bit character set mapping
    ais_charset = '@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_ !"#$%&\'()*+,-./0123456789:;<=>?'
    
    # Convert the binary string to text using 6-bit chunks
    text = ''
    for i in range(0, len(binary_str), 6):
        six_bit_group = binary_str[i:i+6]
        decimal_value = int(six_bit_group, 2)
        text += ais_charset[decimal_value]
    
    # Extract the codes assuming a fixed format: first 10 characters for UN/LOCODEs and next 5 for ERI code
    un_locode_1 = text[:5]
    un_locode_2 = text[5:10]
    eri_code = text[10:15]
    
    # Combine all three codes into a single string
    combined_codes = un_locode_1 + un_locode_2 + eri_code

    print(combined_codes)
    
    return combined_codes
```

Get the Vessel name- and call sign

```python
def get_vessel_and_call(binary_string):
    # AIS 6-bit character set mapping
    ais_charset = '@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_ !"#$%&\'()*+,-./0123456789:;<=>?'

    # Divide binary string into 6-bit groups and convert each to decimal
    name = ''
    for i in range(0, len(binary_string), 6):
        six_bit_group = binary_string[i:i+6]
        decimal_value = int(six_bit_group, 2)
        name += ais_charset[decimal_value]
        
        #Remove any additional @ at the end of the string
        name = name.rstrip('@')

    return name
```

Validate MMSI number

```python
def check_mmsi(mmsi):
    # checks if not null
    try:
        mmsi = str(int(mmsi))
    except:
        return 0, False

    # check if the length is 7 or 9
    if len(mmsi) not in {7, 9}:
        return 0, False

    # should not contain all same digits 000000000
    if len(set(mmsi)) == 1:
        return 0, False

    # should not be consecutive eg: 123456789
    if [int(i) for i in mmsi] == list(range(int(min(mmsi)), int(max(mmsi)) + 1)):
        return 0, False

    return int(mmsi)
```

```python
from datetime import datetime

def decode_ais_message1(binary_str, fields):

    #Columns for all message types
    time_format, packet_type = fields[0].split(' ', 1)
    #time 
    time = datetime.strptime(time_format, "%Y-%m-%dT%H:%M:%S.%fZ")

    #Proprietary packet type
    if packet_type.startswith("$"): 
        return {
            'Timestamp': time,
            'Packet Type': packet_type.lstrip('$'),
            'Prop message 1': fields[1], 
            'Prop message 2': fields[2], 
            'Prop message 3': fields[3], 
            'Prop message 4': fields[4], 
            'Prop message 5': fields[5]
        }

    message_type = int(binary_str[0:6], 2)
    repeat_ind = int(binary_str[6:8], 2)
    mmsi = check_mmsi(int(binary_str[8:38], 2))
    channel = fields[4]

    #Columns different per message type
    ais_version = 0
    imo = 0
    call_sign = 0
    vessel_name = 'NaN'
    ship_type = get_ship_type(0)
    a = 0
    b = 0
    c = 0
    d = 0
    eta = 'NaN'
    draught = 0
    destination = 'NaN'
    nav_status = 'NaN'
    rot = float('nan')
    sog = float('nan')
    cog = 'NaN'
    position_acc = 'NaN'
    long = 0
    lat = 0
    heading = float('nan')
    radio_status = 0
    sotdma = 0
    pos_fix_epfd = get_position_fix_type(0)
    maneuver = get_maneuver_ind(0)

 ################type 1,2,3 ################
    if message_type in [1, 2, 3]:
        #rate of turn
        if len(binary_str) < 50:
            rot = float('nan')
        else:
            rot= int(binary_str[42:50], 2)
            if (1 <= rot <= 126):
                rot = int(rot / 4.733) ** 2
            elif (-126 <= rot <= -1):
                rot = (int(rot / 4.733) ** 2)*-1
            #elif rot == -127 or rot== 127:
                #rot = 'Turn more than 5deg/sec'
            #elif rot == 0:
               # rot = 'Not turning'


        #speed over ground
        if len(binary_str) < 60:
            sog = float('nan')
        else:
            sog = int(binary_str[50:60], 2)*0.1
            if sog == 102.3:
                sog = float('nan')
            elif sog == 102.2:
                sog = '102.2 knots or higher'

        #position accuracy
        if len(binary_str) < 61:
            position_acc = float('nan')
        else:
            position_acc = int(binary_str[60:61], 2)
            if position_acc == 1:
                position_acc = '<10m'
            elif position_acc ==0:
                position_acc = '>10m'


        #longitude
        if len(binary_str) < 89:
            long = float('nan')
        else:
            long = minute_to_dms_long(binary_str[61:89])
            
        #latitude
        if len(binary_str) < 116:
            lat = float('nan')
        else:
            lat = minute_to_dms_lat(binary_str[89:116])

        #Course over ground
        if len(binary_str) < 128:
            cog = float('nan')
        else:
            cog = int(binary_str[116:128], 2)
            if cog == 3600:
                cog = 'NaN'

        #True Heading
        if len(binary_str) < 137:
            heading = float('nan')
        else:
            heading = int(binary_str[128:137], 2)
            if heading == 511:
                heading = float('nan')

        #Radio status
        if len(binary_str) < 168:
            radio_status = float('nan')
        else:
            radio_status = int(binary_str[149:168], 2)

        #Navigation status
        if len(binary_str) < 42:
            nav_status = float('nan')
        else:
            nav_status = get_navigation_status(int(binary_str[38:42], 2))

        #Maneuver
        if len(binary_str) < 42:
            maneuver = get_maneuver_ind(0)
        else:
            maneuver = get_maneuver_ind(int(binary_str[38:42], 2))

    ################type 4 ################

    elif message_type == 4:

        #longitude
        if len(binary_str) < 107:
            long = float('nan')
        else:
            long = minute_to_dms_long(binary_str[79:107])
            
        #latitude
        if len(binary_str) < 134:
            lat = float('nan')
        else:
            lat = minute_to_dms_lat(binary_str[107:134])

        #Type of EPFD
        if len(binary_str) < 138:
            pos_fix_epfd = get_position_fix_type(0)
        else:
            pos_fix_epfd = get_position_fix_type(int(binary_str[134:138], 2))

        #SOTDMA state (radio)
        if len(binary_str) < 168:
            radio_status = float('nan')
        else:
            radio_status = int(binary_str[149:168], 2)

        #ETA 
        if len(binary_str) < 78:
            eta = float('nan')
        else:
            eta = combine_to_datetime(binary_str)


    ###### type 5#####
    elif message_type == 5:

        #AIS version
        if len(binary_str) < 40:
            ais_version = float('nan')
        else:
            ais_version= int(binary_str[38:40],2)

        #IMO number
        if len(binary_str) < 70:
            imo = float('nan')
        else:
            imo= int(binary_str[40:70],2)

        #Call Sign
        if len(binary_str) < 112:
            call_sign = float('nan')
        else:
            call_sign= get_vessel_and_call(binary_str[70:112])

        #Vessel name
        if len(binary_str) < 232:
            vessel_name = 'NaN'
        else:
            vessel_name= get_vessel_and_call(binary_str[112:232])

        #Ship type
        if len(binary_str) < 240:
            ship_type = get_ship_type(100)
        else:
            ship_type= get_ship_type(int(binary_str[232:240],2))

        #Dimension to Bow
        if len(binary_str) < 249:
            a = float('nan')
        else:
            a = int(binary_str[240:249], 2)

        #Dimension to Stern 
        if len(binary_str) < 258:
            b = float('nan')
        else:
            b = int(binary_str[249:258], 2)
            
        #Dimension to Port
        if len(binary_str) < 264:
            d = float('nan')
        else:
            d = int(binary_str[258:264], 2)

        #Dimension to Starboard
        if len(binary_str) < 270:
            c = float('nan')
        else:
            c = int(binary_str[264:270], 2)

        #Position fix type
        if len(binary_str) < 274:
            pos_fix_epfd = get_position_fix_type(0)
        else:
            pos_fix_epfd = get_position_fix_type(int(binary_str[270:274], 2))

        #Draught
        if len(binary_str) < 302:
            draught = float('nan')
        else:
            draught = int(binary_str[294:302], 2)/10

        #Destination
        if len(binary_str) < 422:
            destination = 'NaN'
        else:
            destination = get_destination(binary_str[302:422], 2)

    #Region
    if lat == 0 or long == 0:
        region = 'NaN'
    else: 
        region = get_region(long, lat)

    return {
            'Timestamp': time,
            'Packet Type': packet_type.lstrip('!'),
            'Channel': channel,
            'Message Type': message_type,
            'MMSI': mmsi,
            'Navigation Status': nav_status,
            'Repeat Indicator':repeat_ind,
            'IMO': imo,
            'ROT': rot,
            'SOG': sog,
            'COG': cog,
            'Position Accuracy': position_acc,
            'Longitude':long,
            'Latitude':lat,
            'Region':region,
            'Vessel name': vessel_name,
            'Ship type': ship_type,
            'True Heading': heading, 
            'Radio status': radio_status,
            'Destination': destination, 
            'Maneuver Indicator': maneuver,
            'Draught': draught,
            'Position fix type': pos_fix_epfd,
            'Call sign': call_sign,
            'ETA': eta,
            'A':a, 
            'B': b, 
            'C': c, 
            'D': d
        }
```

Verify_checksum of AIS message

```python
def verify_checksum(nmea_sentence):
    """Verify the NMEA sentence checksum."""
    time, ais_sentence = nmea_sentence.split(' ', 1)
    nmea_data, checksum = ais_sentence.split('*')
    nmea_data = nmea_data.lstrip('!') #removes !
    nmea_data = nmea_data.lstrip('$') #removes $
    

    calc_checksum = 0
    for char in nmea_data:
        calc_checksum ^= ord(char) #returns integer unicode(subset of ASCII)

    return int(checksum, 16) == calc_checksum #converts a checksum that represents a hexadecimal (base-16) number into an integer (base-10)
```

Decode_AIS_nmea (decoding the content of the sentence)

```python
def decode_ais_nmea(nmea_sentence):
    """Decode an AIS NMEA sentence."""
    if not verify_checksum(nmea_sentence):
        return {'Error': 'Checksum mismatch'}

    fields = nmea_sentence.split(',') #fields is array
    payload = fields[5] #payload
    binary_payload = decode_payload(payload)
    decoded_message = decode_ais_message1(binary_payload, fields)

    return decoded_message
```
